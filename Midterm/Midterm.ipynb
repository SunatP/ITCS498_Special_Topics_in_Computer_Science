# Training & Testing Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
factors = ['STATION_1','STATION_2','STATION_3','STATION_4','STATION_5','STATION_6',
          'STATION_7','STATION_8','STATION_9','STATION_10','STATION_11','STATION_12','STATION_13'
          ,'STATION_14','STATION_15','STATION_16','STATION_17','STATION_18','STATION_19','STATION_20',
          'STATION_21','STATION_22','STATION_23','STATION_24','STATION_25','STATION_26','STATION_27','STATION_28'
           ,'STATION_29','STATION_30','STATION_31']
X = data[factors].copy()
Y = data['Fail'].copy()
```

```python
X = StandardScaler().fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 555)
X_train.shape[0] == Y_train.shape[0]
```

```python
regressor = LinearRegression()
regressor.get_params()
regressor.fit(X_train,Y_train)
Y_predicted = regressor.predict(X_train)
mse = mean_squared_error(y_true = Y_train, y_pred = Y_predicted)
# mean_squared_error(y_true, y_pred, squared=False)
print('Mean Square Error : ',mse) # Mean Square Error
print("Root Mean Square Error : ",np.sqrt(mse)) # Root Mean Square Error
```
