```python
url = 'https://pantip.com/search/search/get_search'
keyword = 'covid-19'
topic_id = []
header_search = {
        'accept' : 'application/json',
    'X-Requested-With' : 'XMLHttpRequest'
    
}
param ={'inputtext': keyword,'page': 1}

search_response = requests.post(url,data=param,headers = header_search)
print(search_response.reason)

search_response.encoding = 'utf-8-sig'
search_json = json.loads(search_response.json()['response'])

print(search_json)
```

```python
for i in search_json['data']:
    topic_id.append(i['topic_id'])
print(topic_id)
```

```python
for tid in topic_id:
    time.sleep(5)
    print(tid)
```
