## Get Comment with using Selenium + Chrome Webdriver

```python
from selenium import webdriver

import time

driver=webdriver.Chrome('C:/Users/Sunat/Desktop/Workshop Web Scraping-20200401/chromedriver.exe')

driver.get('https://www.youtube.com/watch?v=' + watch)

driver.execute_script('window.scrollTo(1, 500);')

#now wait let load the comments
time.sleep(5)

driver.execute_script('window.scrollTo(1, 3000);')

comment_user = []
comment_div=driver.find_element_by_xpath('//*[@id="contents"]')
comments=comment_div.find_elements_by_xpath('//*[@id="content-text"]')
for comment in comments:
#     topic_id.append(i['topic_id'])
    print(comment.text)
```

```python
import csv
# import pandas as pd
for comment in comments:
#     topic_id.append(i['topic_id'])
    comment_user.append(comment.text)
    print(comment.text)
#     comment = comment.to_csv('comment.csv', sep='\n')
```

## Get comment to CSV file

```python
import csv
type(comment_user)

with open('output.csv', 'w',encoding="utf-8-sig") as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL,delimiter='\n')
    wr.writerow(['Comment'])
    wr.writerow(comment_user)
```

## Dump comment file to JSON format

```python
import json
with open('data_scrap.json', 'w', encoding='utf-8-sig') as f:
    json.dump(comment_user, f, ensure_ascii=False, indent=4)
```
